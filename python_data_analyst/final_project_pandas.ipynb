```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
#get 2017 data
df_2017 = df[df["Order Date"].dt.year == 2017]
#use .agg and rounded into 3 decimal places
summary_2017 = round(df_2017["Sales"].agg(["sum","mean","std"]),3)
print(summary_2017,'\n')

print(f"This is our 2017 Sales summary\n sum: {summary_2017[0]}\n average: {summary_2017[1]}\n std: {summary_2017[2]}")
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df["Order Date"].dt.year == 2018]
df_2018.groupby("Segment")["Profit"].sum().sort_values(ascending=False).head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 
#define start and end date
start_date = '2019-05-15'
end_date = '2019-12-31'

#filter the date that in between this date by df.between(start,end)
start_end_data = df[df["Order Date"].between(start_date, end_date)]

#Sum of sales group by State then sort ascending order and get top 5 data
round(start_end_data.groupby("State")["Sales"].sum().sort_values().head(5),2)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 

#get the data only West + Central Region and then sum the sales columns
sum_sales_west_central = df[((df["Region"] == "West") | (df["Region"] == "Central"))\
                            & (df["Order Date"].dt.year == 2019)]["Sales"].sum()

#calculated the propotion of west + central (west+central sales / total sales)
proportion_west_centr = (sum_sales_west_central/ df["Sales"].sum()) * 100

print(f"The proportion of total sales in West + Central in 2019 is{proportion_west_centr : .2f}%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
#query 2019-2020 data
df_19_20 = df[(df["Order Date"].dt.year >= 2019) & (df["Order Date"].dt.year <= 2020) ]

#find 10 popular products in terms of number of orders using value_counts() 
top_product_order_19_20 = df_19_20["Product Name"].value_counts().sort_values(ascending = False).head(10)

#find 10 popular products in terms of total sales
top_product_sales_19_20 = df_19_20.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

print("Top order in 2019-2020","\n",top_product_order_19_20,"\n\n")
print("Top sales in 2019-2020","\n",top_product_sales_19_20)
```
